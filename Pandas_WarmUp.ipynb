```python
#### Sorting

df.sort_values(['Attack'], ascending = True)
df.sort_values(['Type 1','Attack'],ascending=[1,0])
```

```python
#### Making changess to the data
#### Create a new column
df['Total'] = df['HP']+df['Attack']+df['Defense']
df['Total 2'] = df.iloc[:,2:5].sum(axis=1) ###column-wise addition

#### Removing a column
df.drop(columns=['Total'])
```

```python
#### changing. the arrangement of the columns
cols = list(df.columns)
print(cols)
df = df[cols[0:4]+[cols[-1]]+cols[4:12]] #### If there is just one column we should consider it as a list
df.head(10)
```

```python
#### Saving your data
df.to_csv("/content/gdrive/My Drive/modified_file.csv",index = False, header=True)### sep could also be used
```

```python
#### Filtering the data
new_df = df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison') & (df['HP']>70)]

#### remove the old index
new_df = new_df.reset_index(drop = True)
## or
new_df.reset_index(drop=True, inplace=True)
new_df.head(5)
```

```python
#### Columns that contain something special
import re ###regular expressions
df.loc[df['Name'].str.contains('Mega')]
df.loc[df['Name'].str.contains('fire|grass',flags=re.I,regex=True)]###has fire or grass in it
df.loc[df['Name'].str.contains('^pi[a-z]*',flags=re.I,regex=True)] ###starts with pi
```

```python
### Conditional Changes

df.loc[df['Type 1'] == 'Fire','Type 1']='New Name'
df.head(10)
```

```python
df.loc[df['Speed']>40,['Generation','Legendary']]=['Test1','Test2']
df
```

```python
##### Group By 
df.groupby(['Type 1']).mean().sort_values('Defense',ascending=False)
df.groupby(['Type 1']).sum()
df['count']=1
df.groupby(['Type 1']).count()['count']
df.groupby(['Type 1','Type 2']).count()['count']
```

```python
#### Wroking with big datasets
new_df = pd.DataFrame (columns = df.columns)
for df in pd.read_csv('/content/gdrive/My Drive/pokemon_data.csv',chunksize=5):
  results = df.groupby(['Type 1']).count()
  new_df = pd.concat([new_df,results])
new_df.head(10)  
```
