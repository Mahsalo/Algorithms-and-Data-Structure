**Lambda, Map, Reduce, Filter**

```python
#### Lambda Function: One line function, no def or return, Lambda parameters: function/return value
mul=lambda x:2*x 
add=lambda a,b: a+b
mx1=lambda a,b:max(a,b)
mx2=lambda a,b:a if a>b else b

print(mul(3),add(3,4),mx1(44,6),mx2(56,90))
```

```python
##### Map function: Apply the same function on each element of a sequence(list...)
##### map(lambda, list), we can used other user defined functions to map
n = [1,2,4,6,7]
print(list(map(lambda x:x*2 , n)))
```

```python
#### Filter function: filters some elements based on the conditions
### we can use the lambda function as a condition

n =[1,4,6,2]
print(list(filter(lambda x:x>2,n)))
```

```python
#### Reduce function: apply a function to each element and then use the results of that
### as the input to the next operation
from functools import reduce
n= [1,2,3,4]
print(reduce(lambda x,y: x+2*y , n)) ### 1+4=5, 5+6=11,  11+8=19
```

**OOP  in Python**


*   Class: A design that can be used to generate real stuff: instances
*   Instances: Real stuff, samples of the class
*   Attributes: features/variables or knowldege that the objects have
*   Methods: Functions
*   Self is the object that we are passing
*   You have to pass the arguments to the "__init__"
*   Everything is an object in Python and objects are stored in the Heap memory of the coomputer
*   **Instance Variables** are defined inside the init function of the class and they depedn on each object separately.
*   **Class/Static Variables** are defined outside of the init function and they are set for all objects. If we change them, they will change for all objects simultanouslty.  
*   **Instance Methods** are defined in an ordinary way.  
*   **Class Methods** are different from **Static Methods**.
*   **Accessor Methods** would access the values in the class.
*   **Mutator Methods** would set/update the values.
*   Instance variables could only work with instance methods.
*   If we are working with the class variables, we need to use the class methods and use **cls** instead of **self**. 
* We need to add a decorator before our method. It starts with @ such as @classmethod or @staticmethod.
* Static methods are good if we want to have a function that is not related to any of the variables in the class (not related to instance or class variables).





```python
class Computer:
  def __init__(self,cpu, ram):
    self.cpu = cpu ### not local variable, they belong to the object so we need self
    self.ram = ram

  def config(self):
    print('cpu is',self.cpu,'\nram is',self.ram)


class Class1:
  def __init__(self):
    self.Name ='mahsa'
    self.age = 31

  def compare(self,other):
    if self.age == other.age and self.Name==other.Name:
      return True
    else:
      return False


comp1 = Computer('i5',8) ###. this is actually the constructor even if we don't define __init__
comp2 = Computer('Ryzen 3',64)

#Computer.config(comp1) ### we are passing the object (self) as the input argument of the config
#Computer.config(comp2)

comp1.config() ### Behind the scenes config takes comp1 as an input object
comp2.config()

#### check the address of each object
print(id(comp1))
print(id(comp2))

#### change the values of each attribute
cl1 = Class1()
cl1.Name = 'Nahid'
print(cl1.Name)
cl2 = Class1()

#### comparing two instances of an object, we need to make our own compare function
if cl1.compare(cl2) == True:
  print('Two objects are the same')
else:
  print('Two objects are different')
```

```python
#### Instance, Class and Static methods.
class Student:
  school = 'faraz'
  def __init__ (self,m1,m2,m3):
    self.m1 = m1
    self.m2 = m2
    self.m3 = m3


  def avg(self):
      return (self.m1 + self.m2 + self.m3)/3

  def get_val(self):
      return self.m1

  def set_val(self,value): ###it gets the value from the user
      self.m1 = value
  @classmethod
  def getSchool(cls):    
      return cls.school

  @staticmethod
  def info():
    print("this method is not relevant to anything special")    
s1 = Student(3,2,66)
s2 = Student(44,3,5)

print(s1.avg())
print(s2.avg())
print(s2.getSchool())
s1.info()



```

**Inheritence**
* If class A is the parent/super-class and class B is the child/sub-class. Class B inherits everything from class A and has access to it.
* **Multi-level** inheritence means that several classes inherit in a long linear way.
* **Multiple** inheritence means that one class could inherit from several non-related classes.
* If the parent class has input variables but the child does not have any inputs, we need to determine the input variables even for the child while definfing. 
* If the child has no __init__ constructor,  it will directly go to the __init__ of the parent but if it has the __init__ itself, it will go to the __init__ of the child. 
* But what if we want to go to both of the constructors: we need to define **super** for the child's constructor.
* If we have multiple inheritence, we will have super and it will go from left to righ (MRO: method resolution order).
* super could be used in front of other methods as well since it means super-class.

3. **Operator Overloading:** change the behavior of pre-defined operatorss ssuch as addition, mmultiplication, subtraction and etc. We need to access the methods by writting def dash dash add dash dash().

4. **Method Overloading:** It means two methods with the same name have different number of parameters/inputs but this concept does not exist in Python. In this case we can have the maximum number of arguments and pass them to the methods and set the initial values of those extra arguments to **None**.

5. **Method Overriding:** Several methods with the same name and same number of arguments and same rturning values but their behaviour is different from one another. If we have two **Func** functions in child and parent with everything the same except for their behaviour, it depends on whether we have called the child or the parent to invoke those functions.  

```python
### Method Overloading 

class Student:
  def __init__(self,m1,m2,m3=None):
    self.m1 = m1
    self.m2 = m2
    self.m3 = m3

  def  addition (self,m1,m2,m3=None):
    if m3 != None:
      return self.m1 + self.m2 + self.m3 
    else:
      return self.m1+self.m2

s1 = Student(1,2,3)
s2 = Student(2,3)
s1.addition(1,2,3)
s2.addition(2,3)   


### Method Overriding:

class BStudent(Student):
  def __init__(self,m1,m2,m3=None):
    self.m1 = m1
    self.m2 = m2
    self.m3 = m3
  def addition(self,m1,m2,m3=None):
    if m3 != None:
      return self.m1 * self.m2 + self.m3 
    else:
      return self.m1*self.m2

bs = BStudent(4,2)
bs.addition(4,2)
```

```python
#### Operator Overloading:
class complex:
  def __init__(self,a,b):
    self.a = a 
    self.b = b

  def __add__(self,OtherObj):  
    return (self.a + OtherObj.a, self.b+OtherObj.b)

s = complex(1,3)
o = complex(4,5)
d = s+o
print(d)
```
