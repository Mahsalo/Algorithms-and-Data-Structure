```python
##### Data structures in Python
#### Sequences : Lists, Strings, Tuples 
#### Sequencing functions

x1 = [1,34,6,2]
x2 = 'Hi there! This is me'
x3 = (34,7)
x4 = (1,5,3,6,3,3)

### Slicing 
print(x1[3],x2[1:3],x3[0],'\n')

### Multiplying
print(x1*2, x2*3, x3*2,'\n')

### Checking the membership (in and not in)
print(6 in x1, "i" not in x2, 34 in x3)

### Iterating through elements
for i in x3:
  print(i)

for i , j in enumerate (x4):
  print(i,j)  

### Count the number. of items
print(x4.count(3),'\n')
print(len(x3),'\n')

### min, max , sum
print(min(x1),max(x2), sum(x4))

### Sorting
print(sorted(x4),sorted(x1),sorted(x2),'\n')

### Find the index of an item (first occurance of an item)
print(x1.index(34),x4.index(3))

### Unpacking
a,b = x3

```

```python
### List functions

### Creating the new list and list comprehension
x=list()
x=[1,4,5]
x=list((3,4))
x1 = [11,45,7,2.67,8,110,90,4]
print(x)

#### List comprehension: creating new list by applying functions on each element of the previous list
x = [i*7 for i in x if i!=4]
print(x)

### Deleting items
del(x1[0])
print(x1)

### Appending items to the list
x.append(46)
print(x)

### Inserting to a specific position in the list
x.insert(1,['a','b'])
print(x)

### pop and remove
x1.pop()
x1.remove(7) ### remove values
print(x1)

### Reverse: str does not have the reverse function it's just for the lists
x.reverse()
print(x)

### Sort (in-place) vs Sorted (not in-place)
x1.sort()
print(x1)
```

```python
#### Tuples are immutable (don't change or update or assign values)
x = ()
x = (1,2,54)
x = 4,5,7 ## no need for paranthesis
x = 4, ### single item tuple
x = tuple(x1) #### change the list to a tuple

print(x)

#### Important: if there is a list inside a tuple, we can change the list but otherwise we cannot make any changes to the tuple
x = ([1,5,7],2,3)
###x[0] = 5 ###error
x[0][1]=20
print(x)
```

```python
#### Sets: no duplicates found in sets, no order, sets do not have slicing
x = {3,5,2,5,2,4}
x =set()
x = set(x1)

#### set comprehension
x2 ={4*i for i in x if i!=3}
print(x,x2)

#### Set operations
x.add(444)
x.pop()
print(len(x))
###x.clear()


#### Standard math operations
x1 = {3,4,7,23,4}
x2 ={44 , 7 ,32,22}
#Intersection
print(x1 & x2)
#Union
print(x1 | x2)
#XOR not in 1 or not in 2: Everything except for the intersection
print(x1 ^ x2)
#Difference: In 1 but not in 2
print(x1-x2)
#Subset and superset
print(x1<=x2)
```

```python
#### Dictionary: Key-Value pairs, random order
x1 = {'pork':2.3, 'beef':4, 'chicken':6.5}
x = dict()
x2 = dict([('pork',22),('beef',5),('chicken',56)]) ### dict could could be a list of tuples of pairs
x3 =dict(pork=2.4) 

#### Dict. operations
x1['pork'] = 44 ### changing the values
x1['lamb'] = 56 ### adding new keys
del x1['beef']
print(x1)
## x1.clear()
## del x1

print(x1.keys())
print(x1.values())
print(x1.items())

### Iterating a dictionary
for k,v in x1.items():
  print(k,v)
```

```python
#### Lambda Function: One line function, no def or return, Lambda parameters: function/return value
mul=lambda x:2*x 
add=lambda a,b: a+b
mx1=lambda a,b:max(a,b)
mx2=lambda a,b:a if a>b else b

print(mul(3),add(3,4),mx1(44,6),mx2(56,90))
```

```python
##### Map function: Apply the same function on each element of a sequence(list...)
##### map(lambda, list), we can used other user defined functions to map
n = [1,2,4,6,7]
print(list(map(lambda x:x*2 , n)))
```

```python
#### Filter function: filters some elements based on the conditions
### we can use the lambda function as a condition

n =[1,4,6,2]
print(list(filter(lambda x:x>2,n)))
```

```python
#### Reduce function: apply a function to each element and then use the results of that
### as the input to the next operation
from functools import reduce
n= [1,2,3,4]
print(reduce(lambda x,y: x+2*y , n)) ### 1+4=5, 5+6=11,  11+8=19
```
